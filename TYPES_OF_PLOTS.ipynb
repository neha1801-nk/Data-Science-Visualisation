```python
# ============================================
# A. UNIVARIATE PLOTS
# ============================================
#NEHA.N.K | 24BAD078
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset
df = pd.read_csv("happiness_2017.csv")

# 1. Histogram (Matplotlib)
plt.hist(df["HappinessScore"], bins=10)
plt.xlabel("Happiness Score")
plt.ylabel("Frequency")
plt.title("Histogram of Happiness Score")
plt.show()

# 2. Histogram (Seaborn)
sns.histplot(df["HappinessScore"], bins=10, kde=True)
plt.title("Histogram of Happiness Score (Seaborn)")
plt.show()


# 3. Boxplot (Matplotlib)
plt.boxplot(df["HappinessScore"])
plt.ylabel("Happiness Score")
plt.title("Boxplot of Happiness Score")
plt.show()

# 4. Boxplot (Seaborn)
sns.boxplot(y=df["HappinessScore"])
plt.title("Boxplot of Happiness Score (Seaborn)")
plt.show()


# 5. Violin Plot
sns.violinplot(y=df["HappinessScore"])
plt.title("Violin Plot of Happiness Score")
plt.show()


# 6. Countplot (Categorical)
sns.countplot(y="Region", data=df)
plt.title("Count of Countries by Region")
plt.show()
```

```python
# ============================================
# B. BIVARIATE PLOTS
# ============================================

#NEHA.N.K | 24BAD078
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset
df = pd.read_csv("happiness_2017.csv")

# 8. Scatter Plot (Numeric vs Numeric)
sns.scatterplot(
    x="Log GDP per capita",
    y="HappinessScore",
    data=df
)
plt.title("Log GDP per Capita vs Happiness Score")
plt.show()


# 9. Line Plot (Trend)
sns.lineplot(
    x=df.index,
    y="HappinessScore",
    data=df
)
plt.xlabel("Index")
plt.ylabel("Happiness Score")
plt.title("Trend of Happiness Score")
plt.show()


# 10. Bar Plot (Top 10 Happiest Countries)
top10 = df.sort_values("HappinessScore", ascending=False).head(10)

sns.barplot(
    x="HappinessScore",
    y="Country",
    data=top10
)
plt.title("Top 10 Happiest Countries")
plt.show()


# 11. Strip Plot (Categorical vs Numeric)
sns.stripplot(
    x="Region",
    y="HappinessScore",
    data=df,
    jitter=True
)
plt.xticks(rotation=90)
plt.title("Strip Plot of Happiness Score by Region")
plt.show()


# 12. Swarm Plot (Categorical vs Numeric)
sns.swarmplot(
    x="Region",
    y="HappinessScore",
    data=df
)
plt.xticks(rotation=90)
plt.title("Swarm Plot of Happiness Score by Region")
plt.show()
```

```python
# ============================================
# C. MULTIVARIATE PLOTS
# ============================================
# NEHA.N.K | 24BAD078
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("happiness_2017.csv")


# 13. Pairplot (Numeric variables)
sns.pairplot(df.select_dtypes(include=np.number))
plt.show()


# 14. Jointplot (Two numeric variables)
sns.jointplot(
    x="Log GDP per capita",
    y="HappinessScore",
    data=df
)
plt.show()


# 15. Heatmap (Correlation Matrix)
corr = df.select_dtypes(include=np.number).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()


# 16. Clustermap (Hierarchical clustering)
sns.clustermap(
    corr,
    cmap="coolwarm",
    annot=True
)
plt.show()
```
